```python
# !pip install openai

import os
import pandas as pd
from tqdm import tqdm
from openai import OpenAI
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/mohitgupta-omg/Kaggle-SMS-Spam-Collection-Dataset-/master/spam.csv', encoding='latin-1')

data.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], axis=1, inplace=True)

data.columns = ['label', 'text']

data.head()
```

```python
client = OpenAI()

predict = []
for i in tqdm(range(500)):
    completion = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": "Я передам тебе текст, я хочу что бы ты отнес его к одному из двух классов: spam или ham и вывел мне получившийся класс. Например: Ok lar... Joking wif u oni... -> ham. Free entry in 2 a wkly comp to win FA Cup final tkts 21st May 2005. Text FA to 87121 to receive entry question(std txt rate)T&C's apply 08452810075over18's -> spam"},
            {
                "role": "user",
                "content": data["text"][i]
            }
        ]
    )
    predict.append(completion.choices[0].message.content)
```

```python
accuracy_score(data['label'][:499], predict)
```
